## Derivatives

Derivatives tell the slope of a function, useful to measure the rate of change at any point

often used in machine learning - especially gradient descent

```python
x = sp.symbols('x')
```

```python
f = x**2
```

```python
sp.plot(f)
```

How steep is the curve at *x* = 2.5

we measure the steepness with the **tangent line**. Where the straight line just touches the point at 2.5
this can be approximated by creating a line intersecting 2.5 and a really close number to 2.5 (i.e 2.6)

```python
f.subs(x,2.5)
```

```python
f.subs(x,2.6)
```

```python
slope = (6.76 - 6.25) / (2.6 - 2.5)
slope
```

```python
f.subs(x,2.500000001)
```

```python
slope = (6.250000005 - 6.25) / (2.6 - 2.5)
slope
```

```python
def derivative_x(f, x, step_size):
    """
    Slope = m
    f = or the shape of the line
    x = point
    step_size = close point
    
    m = (function(point + close point) - function(point) / (point + close point) - point
     this is the same as m = (y2 -y1) / (x2 - x1)
    
    """
    m = (f(x + step_size) - f(x)) / ((x + step_size) -x)
    return m
```

```python
def my_function(x):
    return x ** 2
```

```python
slope_at_x = derivative_x(my_function, 2.5, .0000000001)
slope_at_x
```

d/dx indicates a derivative with respect to x

```python
x = sp.symbols('x')
```

```python
f = x**2
```

```python
dx_f = sp.diff(f)
dx_f
```

```python
slope_at_x = dx_f.subs(x,2.0)
slope_at_x
```
